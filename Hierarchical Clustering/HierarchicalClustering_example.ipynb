* Gözlemler daha fazla sayıda alt kümelere ayrılmak isteniyorsa k-means yerine hiyerarşik kümeleme kullanılabilir.
* K-meanste 3-4 kümeye ayrılır, burada ise bu **kümelerin altında da kümeler oluşturabilriz.**
* **Hiyerarşik kümelemede hem gözlemler hem de değişkenler kümeleme işlemine sokulabilir.** Fakat K-means ile sadece gözlemlerin kümelenmesi mümkündür.
* Hiyerarşik yöntemlerde **küme sayısına dendogram sonuçlarına bakılarak karar verilir**, hiyerarşik olmayanlarda ise uygulama yapılmadan önce belilenir.
* Karar ağaçlarına benzerlik soru işareti yaratabilir, farkı: karar ağaçlarında ayırma işlemi hedef değişkene göre yapılırken **hiyerarşik kümelemede bağımlı değişken olmadıgı için gözlemler bağımsız değişkenler üzerinden yapılan uzaklık hesaplarına göre kümelere ayrılır.**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import warnings 
warnings.filterwarnings("ignore", category=DeprecationWarning)
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)

df = pd.read_csv("../USArrests.csv", index_col = 0)
df.head()
```

```python
from scipy.cluster.hierarchy import linkage
```

```python
hc_complete = linkage(df, "complete")
hc_average = linkage(df, "average")
```

```python
from scipy.cluster.hierarchy import dendrogram
```

```python
plt.figure(figsize = (15,10))
plt.title("Hiyerarşik kümeleme dendogram")
plt.xlabel("Gözlem birimleri")
plt.ylabel("Uzaklıklar")
dendrogram(hc_complete, leaf_font_size = 10);
```

```python
plt.figure(figsize = (15,10))
plt.title("Hiyerarşik kümeleme dendogram")
plt.xlabel("Gözlem birimleri")
plt.ylabel("Uzaklıklar")
dendrogram(hc_average, leaf_font_size = 10);
```

```python
plt.figure(figsize = (15,10))
plt.title("Hiyerarşik kümeleme dendogram")
plt.xlabel("Gözlem birimleri")
plt.ylabel("Uzaklıklar")
dendrogram(hc_complete, truncate_mode ="lastp", p = 10, show_contracted= True, leaf_font_size = 10);
```
